#Frugal Streaming for Estimating Quantiles

```python
%matplotlib inline
from ggplot import *
import pandas as pd
import numpy as np
```

```python
stream = np.random.normal(loc=100, scale=10, size=10000)
```

```python
df = pd.DataFrame({"x": stream})

ggplot(df, aes(x='x')) \
+ geom_histogram(binwidth=1, alpha=0.6) \
+ theme_seaborn(style='whitegrid', context='notebook', gridweight ='light')
```

```python
median = np.percentile(stream, 50)
median
```

```python
def frugal_1u_median(stream):
    m = 0
    trace = []
    for item in stream:
        if item > m:
            m += 1
        elif item < m:
            m -= 1
        trace.append(m)
    return m, trace
```

```python
m, trace = frugal_1u_median(stream)
m
```

```python
df = pd.DataFrame({'Frugal-1U-Median': trace, 'Median': median})

df['x'] = df.index
df = pd.melt(df, id_vars='x')

ggplot(df, aes(x="x", y="value", color='variable')) \
+ geom_line(size=3, alpha=0.6) \
+ xlim(0, 300) \
+ theme_seaborn(style='whitegrid', context='notebook', gridweight ='light') \
+ labs("iterations", "median value")
```

```python
quantile095 = np.percentile(stream, 95)
quantile095
```

```python
def frugal_1u(stream, q):
    m = 0
    trace = []
    for item in stream:
        r = np.random.random()
        if item > m and r > 1 - q:
            m += 1
        elif item < m and r > q:
            m -= 1
        trace.append(m)
    return m, trace
```

```python
m, trace = frugal_1u(stream, 0.95)
m
```

```python
df = pd.DataFrame({'Frugal-1U-Median': trace, '0,95-Quantile': quantile095})

df['x'] = df.index
df = pd.melt(df, id_vars='x')

ggplot(df, aes(x="x", y="value", color='variable')) \
+ geom_line(size=3, alpha=0.6) \
+ xlim(0, 300) \
+ theme_seaborn(style='whitegrid', context='notebook', gridweight ='light') \
+ labs("iterations", "0,95-quantile value")
```

```python
def frugal_2u(stream, q, f):
    m = 0
    step = 1
    sign = 1
    trace = []
    
    for item in stream:
        if item > m and np.random.random() > 1 - q:
            # Increment the step size if and only if the estimate keeps moving in
            # the same direction. Step size is incremented by the result of applying
            # the specified step function to the previous step size.
            step += f(step) if sign > 0 else -1 * f(step)
            # Increment the estimate by step size if step is positive. Otherwise,
            # increment the step size by one.
            m += step if step > 0 else 1
            # Mark that the estimate increased this step
            sign = 1
            # If the estimate overshot the item in the stream, pull the estimate back
            # and re-adjust the step size.
            if m > item:
                step += (item - m)
                m = item
        # If the item is less than the stream, follow all of the same steps as
        # above, with signs reversed.
        elif item < m and np.random.random() > q:
            step += f(step) if sign < 0 else -1 * f(step)
            m -= step if step > 0 else 1
            sign = -1
            if m < item:
                step += (m - item)
                m = item
        # Damp down the step size to avoid oscillation.
        if (m - item) * sign < 0 and step > 1:
            step = 1
        trace.append(m)
    return m, trace
```

```python
_, trace2_1 = frugal_2u(stream, 0.95, f=lambda x: 1.0)
_, trace2_2 = frugal_2u(stream, 0.95, f=lambda x: 2.0)
_, trace2_10 = frugal_2u(stream, 0.95, f=lambda x: 10.0)
_, trace2_x = frugal_2u(stream, 0.95, f=lambda x: x)
```

```python
df = pd.DataFrame({'Frugal-1U': trace, 'Frugal-2U: f(x) = 1': trace2_1, 'Frugal-2U: f(x) = 2': trace2_2, \
                   'Frugal-2U: f(x) = 10': trace2_10, 'Frugal-2U: f(x) = x': trace2_x, 'quantile': quantile095})

df['x'] = df.index
df = pd.melt(df, id_vars='x')

ggplot(aes(x='x', y='value', color='variable'), df) \
+ geom_line(size=1) \
+ xlim(0, 300) \
+ theme_seaborn(style='whitegrid', context='notebook', gridweight ='light') \
+ labs('iterations', '0,95-quantile value')
```
