```python
from matplotlib import pyplot
from sklearn import datasets
from sklearn.cluster import KMeans
```

```python
X, Y = datasets.make_blobs(centers=4, cluster_std=0.6, random_state=0)
pyplot.scatter(X[:,0], X[:,1])
```

```python
pyplot.scatter(X[:,0], X[:,1], c=Y)
```

```python
kmeans = KMeans(4, random_state=1)
clusters_kmeans = kmeans.fit(X)
pyplot.scatter(X[:,0], X[:,1], c=clusters_kmeans.labels_)
```

```python
centroids = clusters_kmeans.cluster_centers_
pyplot.scatter(centroids[:,0], centroids[:,1], 100)
pyplot.scatter(X[:,0], X[:,1], c=clusters_kmeans.labels_)
pyplot.scatter(X[:,0], X[:,1], c=clusters_kmeans.labels_)
```
